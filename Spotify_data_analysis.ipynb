```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('/content/tracks.csv')
df_tracks.head()
```

```python
#null values

pd.isnull(df_tracks).sum()
```

```python
 df_tracks.info()
```

```python
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks.query('popularity > 70', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks.index = pd.to_datetime(df_tracks.index, errors='coerce', format='mixed')
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(['key','mode','explicit', 'id', 'name', 'artists', 'id_artists', 'release_date'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g',vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='loudness', x='energy', color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', color='r').set(title='Popularity vs Acousticness Correlation')
```

```python
df_tracks['dates']=df_tracks['release_date']
df_tracks.dates=pd.to_datetime(df_tracks.dates, format='mixed')
years=df_tracks.dates.dt.year
```

```python
#pip install --user seaborn==0.11.0
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Distibution of Tracks over the Years')
```

```python
total_dr = df_tracks.duration
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation = 90)
```

```python
total_dr = df_tracks.duration
sns.set_style(style = "whitegrid")
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title="Year vs Duration")
plt.xticks(rotation = 60)
```

```python
df_genre=pd.read_csv('/content/SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title('Duration of the songs in different genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
famous = df_genre.sort_values('popularity', ascending = False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title="popular genres on Spotify")
```
